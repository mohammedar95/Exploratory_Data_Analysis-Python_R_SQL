### Part 1: Filter Data and Fill NA

1.   The Winter Olympics have only been around since 1924 while the Summer Olympics started in 1896.  Let's filter this dataframe to `Year >= 1924` so that we have an even playing field within which to compare Winter performance to Summer performance.
2.   The `Medal` column contains many `NaN` values.  Use `fillna` to substitute these values with `"None"` (i.e. the string "None") instead.


```python
# Write code below
#1

#prepare the dataset for summer and winter comparsion and set it to year >= 1924
medal_athlete = athlete[(athlete['Year'] >= 1924)].sort_values(by='Year')
medal_athlete
```

```python
#2
# replace NaN with None 
medal_athlete["Medal"].fillna('None', inplace=True)
```

```python
medal_athlete
```

### Part 2: Calculate Weighted Medal Points

When evaluating country performance in the Olympics, we don't want to count Gold medals as equivalent to Silver/Bronze medals.  Let's assign "Medal Points" to each tier of Olympic medal as follows:

*   Gold:  3 points
*   Silver: 2 points
*   Bronze: 1 point
*   No medal: 0 points

Write a function `medal_to_points` to transform and create a new column called `MedalPoints`.  `medal_to_points` will accept a string (possible values are `"Gold"`, `"Silver"`, `"Bronze"`, `"None"`) and returns the appropriate number of points as an integer.

Then add a new column `MedalPoints` to your dataframe using `transform` and the newly defined function `medal_to_points`

```python
medal_athlete
```

```python
# Define medal_to_points below
def medal_to_points(medal_name):
  medal = str(medal_name)
  # Insert logic to convert medal to point value
  if medal == 'Gold':
    return 3
  elif medal == "Silver":
    return 2
  elif medal == "Bronze":
    return 1
  else:
    return 0
    # Return the converted point value
    return medal_name
```

```python
#adding a new column with the new metric points
medal_athlete['MedalPoints'] = medal_athlete.Medal.transform(medal_to_points)
```

```python
medal_athlete
```

### Part 3: Aggregate MedalPoints by Country and Season

We need to figure out how each country performed in the Summer, and how each country performed in the winter.  Now that we have created the performance metric `MedalPoints`, we need to aggregate by Country and Season to figure out how many `MedalPoints` each country earned in each season (Summer and Winter).

1.   Using `NOC` for country and `Season` for season, group by these two fields and `sum()` the field `MedalPoints` to determine how many "Medal Points" each country earned in each season (Summer vs Winter)
2.   Store the results of the above in a new dataframe



```python
# Write your code below

#create a variable and groupby country and season and then sum all medalpoints
medal_by_points = medal_athlete.groupby(by=['NOC', 'Season'], as_index=False )['MedalPoints'].sum()
```

```python
medal_by_points.head(n=30)
```

### Part 4: Calculate the % of Medal Pts earned in each Season for each Country

To determine which countries perform better in Winter as compared to Summer, let's rank them by the % of Medal Points earned in the Winter.  In order to do this, we need to calculate a % of total medal points earned by season (Summer vs Winter).

1.   Using your new dataframe, group by `NOC` (Country) only and transform `MedalPoints` into a % of total using a function (can be a lambda function). **Note: if stuck here, see lecture and lecture notes where we did a similar transformation** 
2.   Add the series generated above as a new column `PctTotal` to the dataframe.  



```python
# groupby country and transform medalpoints to calculte summer/summer+winter
medal_by_points['PctTotal'] = medal_by_points.groupby(by='NOC', as_index=False)['MedalPoints'].transform(lambda x: x/x.sum())
```

### Part 5: Report Results

For countries **earning at least 100 MedalPoints in the Winter Olympics**, list the top 3 countries that earned the highest % of their Medal Points in the Winter Olympics.  Also list the percentage of Medal Points earned and total number of Medal Points earned in the Winter Olympics.

```python
# Write your code below
# filter the dataframe to only winter
winter_ranking = medal_by_points[(medal_by_points.Season == 'Winter')] 
#filter to only above 100 medalpoints
top_winter = winter_ranking[(winter_ranking.MedalPoints >= 100)]
# sort the result to highest to get the top 3
top_winter.sort_values(by='PctTotal', ascending=False)
```
